## Quick Linear Regression!

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,20))
y = x * 2 -1 

print(x)
print(y)
```

```
# 1번 청소 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언
model = keras.models.Sequential()

# 3번 모델 블록 조립
model.add( keras.layers.Input(shape=(1,)) )
model.add( keras.layers.Dense(1) )

## 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(1,)) )

# 4번 컴파일 
model.compile(loss='mse',
              optimizer='adam')
# loss : 내가 생성한 모델과 실제 값을 무슨 방법으로 비교할래 
# optimizer : gradient decent를 어떤 방식으로 진행할 것 인가
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
print(y)
```

```
model.predict(x)[:5].reshape(-1) # 가로 형태로 보기
```

```
import tensorflow as tf
import tensorflow.keras as keras

import numpy as np
```

```
x = np.array(range(0,20)) 
y = np.array([0]*10 + [1]*10)
print(x)
print(y)
```

```
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# model에 인풋 값을 받는 레이어를 넣음
model.add( keras.layers.Input(shape=(1,)) )
# model에 Dense 레이어를 넣을거야 (최초의 레이어) : weight를 곱하고, bias를 더해주는 과정
model.add( keras.layers.Dense(1, activation='sigmoid') )
# 이진분류에 맞게 값이 나옴


# 오리지널 Sequential API
# model.add( keras.layers.Dense(1, input_shape=(1,), activation='sigmoid') )

# 컴파일 해주렴
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics=['accuracy'])
```

```
# 데이터를 넣어서 학습시키자!
model.fit(x[:15], y[:15], epochs=10, verbose=1)

# 결과 출력해줘!
print(y[15:])
print(model.predict(x[15:]))
```

## Now, Your turn!

```
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,40)) 
y = np.array([0]*20 + [1]*20)
print(x)
print(y)
```

```
keras.backend.clear_session()# 메모리 문제 때문에 그렇다. 안그러면 남아서 _1. _2로 모델들이 추가된다.

model = keras.models.Sequential()

model.add( keras.layers.Input(shape=(1,)) )
model.add(keras.layers.Dense(1, activation='sigmoid'))# 컴파일 해주렴 

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])# loss를 보고 자동으로 판단해주는 정확도 
# metrics=keras.metrics.BinaryAccuracy(threshold=지정가능)
# 분포와 분포를 비교하는데 사용하는데 사용하는 crossentropy를 쓰는데 직관적이진 않다.
# 정확도를 넣어서 비교하기 쉽게 확인할 수 있다.
```

```
model.fit(x[:15], y[:15], epochs=10, verbose=1)# 이진분류는 0.5보다 클 때 1 작으면 0이다. 그래서 예측이 떨어진다.
```

```
# 결과 출력해줘!
print(y[15:])
print(model.predict(x[15:]))
```
