## Quick Linear Regression!

```
import tensorflow as tf
from tensorflow import keras
# import keras : 외부 keras, 원래는 따로 운영이 되어있었는데 개발자가 구글에 들어가게 되어서 통합됨 위의 걸로 쓰면 됨
# 현업에서는 keras 더 많이 사용 함 , 연구실은 pytorch
import numpy as np
```

```
x = np.array(range(0,20)) 
y = x * 2 -1

print(x)
print(y)
```

```
# 자주 해줘야 함!
x.shape, y.shape
# input layer 안에 x라는 값을 하나 가지고 있다고 이해
# x가 데이터 프레임 하나의 feature라고 생각
```

```
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# 레이어들을 사슬로 연결하 듯이 연결!
input_layer = keras.layers.Input(shape=(1,))
output_layer = keras.layers.Dense(1)(input_layer)# 사슬처럼 연결해주어야 함

# 모델의 시작과 끝을 지정
model = keras.models.Model(inputs=input_layer, outputs=output_layer)

# 컴파일 해주렴
model.compile(loss = 'mse', optimizer = 'adam')
```

```
# 데이터를 넣어서 학습시키자!
model.fit(x, y, epochs=10, verbose=1)
```

```
# 결과 출력해줘!
print(y)
print(model.predict(x).reshape(-1,) )
```

## Quick Logistic Regression!

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(0,20)) 
y = np.array([0]*10 + [1]*10)

print(x)
print(y)
```

```
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# 레이어들을 사슬로 연결하 듯이 연결!
input_layer = keras.layers.Input(shape=(1,))
output_layer = keras.layers.Dense(1, activation='sigmoid')(input_layer)

# 모델의 시작과 끝을 지정
model = keras.models.Model(inputs=input_layer, outputs=output_layer)


# 컴파일 해주렴
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```
# 데이터를 넣어서 학습시키자!
model.fit(x, y, epochs=10, verbose=1)

# 결과 출력해줘!
print(y)
print(model.predict(x).reshape(-1,) )
```
