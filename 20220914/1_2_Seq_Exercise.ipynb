# Keras로 Linear&Logistic Regression 조금 더 맛보기

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
```

# Linear Regression on 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
from tensorflow.python import metrics
keras.backend.clear_session()

model = keras.models.Sequential()

model.add(keras.layers.Input(shape=(13, )))
model.add(keras.layers.Dense(13))

model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(x, y, epochs=100, verbose=1)
```

```
model.predict(x)[:15].reshape(-1)
```

```
y[:15]
```

# Logistic Regression 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

# 모양 확인해볼 것
x.shape, y.shape
```

```
keras.backend.clear_session()
model = keras.models.Sequential()
model.add(keras.layers.Input(shape=(30,)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x, y, epochs=100, verbose=1)
```

```
model.predict(x)[:15].reshape(-1)
```

```
y[:15]
```
