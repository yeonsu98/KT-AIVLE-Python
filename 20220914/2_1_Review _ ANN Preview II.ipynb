```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target
```

```python
x.shape, y.shape
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2022)
```

```python
#####################
# 라이브러리 불러오기
#####################

import tensorflow as tf
from tensorflow import keras
```

```python
########
# 모델링
########

keras.backend.clear_session()

il = keras.layers.Input(shape=(30,), name='inputLayer')
ol = keras.layers.Dense(1, activation='sigmoid', name='outputLayer')(il)

model = keras.models.Model(il, ol)

model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.summary()
# functional은 input 레이어가 표현 됨
# sequancial은 다음 레이어에 포함되어 표현
# param = 30+1
```

```python
###########
# 모델 학습
###########

model.fit(x_train, y_train, epochs=100, verbose=1)
```

```python
###########
# 모델 예측
###########

y_pred = model.predict(x_test)

print(y_pred[:15])
print(y_test[:15])
```

# 히든 레이어 추가하여 모델링!

    히든 레이어 2개를 추가하여 다시 모델링!
    히든 레이어 노드 수 : 32
    히든 레이어 activation = 'relu'

```python
########
# 모델링
########

keras.backend.clear_session()

il = keras.layers.Input(shape=(30,), name='inputLayer')
hl1 = keras.layers.Dense(32, activation='relu')(il)
hl2 = keras.layers.Dense(32, activation='relu')(hl1)
ol = keras.layers.Dense(1, activation='sigmoid', name='outputLayer')(hl2)

model = keras.models.Model(il, ol)

model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.summary()
```

```python
###########
# 모델 학습
###########

model.fit(x_train, y_train, epochs=100, verbose=1)
```

```python
###########
# 모델 예측
###########

import numpy as np

y_pred = model.predict(x_test)
y_pred = np.where(y_pred >= 0.5, 1, 0)

print(y_pred.reshape(-1))
print(y_test)
```
