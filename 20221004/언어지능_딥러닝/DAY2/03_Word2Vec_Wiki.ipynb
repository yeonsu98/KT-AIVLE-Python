```python
from __future__ import print_function

import codecs
import os
import gensim
```

```python
%%time
def read_data(filename):
    with codecs.open(filename, 'r', 'utf-8') as f:
        data = [line.split(' ') for line in f.read().splitlines()]
    return data
        
data = read_data('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki/wiki.txt')

print(len(data))
print(data[0])
```

```python
%%time
import gensim
# Word2Vec 생성
# sg=1이면 skip-gram 0이면 cbow 방식 사용, min_count 단어 빈도수, size 벡터차원
model = gensim.models.Word2Vec(data, min_count=1, size=100, window=5, sg=1, seed=10)
model.save('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki/wiki.w2v')
```

```python
%%time
model = gensim.models.Word2Vec.load(os.path.join('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki', 'wiki.w2v'))
```

```python
model.wv['고양이']
```

```python
model.wv.similarity('고양이', '개')
```

```python
for t in model.wv.most_similar('강아지'):
    print(t[0])
```

```python
result = model.wv.accuracy(os.path.join('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki','word_analogy_korean.txt')) 
for r in result:
    print('%s\t%s\t%s' % (r['section'], len(r['correct']), len(r['incorrect'])))
```

```python
result = model.wv.accuracy(os.path.join('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki','questions-words-kor.txt')) 
for r in result:
    print('%s\t%s\t%s' % (r['section'], len(r['correct']), len(r['incorrect'])))
```
