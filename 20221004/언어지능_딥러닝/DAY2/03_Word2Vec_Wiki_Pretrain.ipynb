```python
model = gensim.models.Word2Vec.load(os.path.join('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki', 'wiki.w2v'))
```

```python
model.wv["고양이"]
# "고양이"의 word vector
```

```python
# Word Analogy Test (임베딩된 단어벡터들이 의미론적/문법적 관계를 잘 반영하는지 테스트)
# Find the top-N most similar words. Positive words contribute positively towards the similarity, negative words negatively.
for t in model.wv.most_similar(positive=["파리", "일본"], negative=["도쿄"], topn=10):
    print('%s\t%f' % (t[0], t[1]))
```

```python
for t in model.wv.most_similar(positive=["앙카라", "네팔"], negative=["카트만두"], topn=20):
    print('%s\t%f' % (t[0], t[1]))
```

```python
for t in model.wv.most_similar(positive=["베를린", "영국"], negative=["런던"], topn=20):
    print('%s\t%f' % (t[0], t[1]))
```

```python
result = model.wv.accuracy(os.path.join('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki','word_analogy_korean.txt'))
for r in result:
    print('%s\t%s\t%s' % (r['section'], len(r['correct']), len(r['incorrect'])))
```

```python
result = model.wv.accuracy(os.path.join('/content/gdrive/My Drive/Colab Notebooks/aivle/data/wiki','questions-words-kor.txt'))
for r in result:
    print('%s\t%s\t%s' % (r['section'], len(r['correct']), len(r['incorrect'])))
```
